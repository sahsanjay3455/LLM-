```python
from langgraph.graph import StateGraph , START,END
from langchain_openai import ChatOpenAI
from dotenv import  load_dotenv
from typing import  TypedDict
```

```python
load_dotenv()

model = ChatOpenAI(
    model="stepfun/step-3.5-flash:free",
    base_url="https://openrouter.ai/api/v1",
    temperature=0.3,
)
```

```python
class llm_state(TypedDict):
    topic:str
    outline:str
    blog:str
    score:str
    
    
```

```python
def generate_outline(state:llm_state)->llm_state:
    topic=state['topic']
    prompt=f"you need to generate outline on the topic :{topic}"
    response=model.invoke(prompt).content
    state['outline']=response

    return state
```

```python
def generate_blog(state:llm_state)->llm_state:
    topic=state['topic']
    outline=state['outline']
    prompt=f"you need to generate blog on the topic :{topic} and according to the outline :{outline}"
    
    response=model.invoke(prompt).content
    state['blog']=response

    return state
```

```python
def generate_score(state:llm_state)->llm_state:
    outline=state['outline']
    blog=state['blog']
    prompt=f"based on the this  outline:{outline} read my blog:{blog} and generate the integer score out of 5  "
    response=model.invoke(prompt).content
    state['score']=response

    return state
```

```python
#define graph
graph=StateGraph(llm_state)

#define node
graph.add_node('generate_outline',generate_outline)
graph.add_node('generate_blog',generate_blog)
graph.add_node('generate_score',generate_score)
#define edges
graph.add_edge(START,'generate_outline')
graph.add_edge('generate_outline','generate_blog')
graph.add_edge('generate_blog','generate_score')
graph.add_edge('generate_score',END)
#compile graph
llm_workflow=graph.compile()
```

```python
intial_state={"topic":"Rise of AI in India"}
final_state=llm_workflow.invoke(intial_state)
```
